```python
def OHE(data,column_name):
    one_hot_encoded_data = pd.get_dummies(data[column_name],prefix=column_name)
    data=pd.concat([data,one_hot_encoded_data],axis=1)
    data.drop(column_name,axis=1,inplace=True)
    del one_hot_encoded_data
    return data
```
